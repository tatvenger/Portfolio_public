## Описание проекта <a class="anchor" id="Описание проекта"></a>

Чтобы оптимизировать производственные расходы, металлургический комбинат «Стальная птица» решил уменьшить потребление электроэнергии на этапе обработки стали. Для этого комбинату нужно контролировать температуру сплава.

**Цель проекта**\
Необходимо построить модель машинного обучения, предсказывающую температуру сплава.
- Целевой признак **target** - последняя температура партии сплава.
- Целевая метрика - MAE (необходимо получить менее 6.8 на тестовой выборке).


**Описание данных**\
Данные состоят из нескольких файлов, полученных из разных источников:
- `data_arc_new.csv` — данные об электродах;
- `data_bulk_new.csv` — данные о подаче сыпучих материалов (объём);
- `data_bulk_time_new.csv` — данные о подаче сыпучих материалов (время);
- `data_gas_new.csv` — данные о продувке сплава газом;
- `data_temp_new.csv` — результаты измерения температуры;
- `data_wire_new.csv` — данные о проволочных материалах (объём);
- `data_wire_time_new.csv` — данные о проволочных материалах (время).

**Файл `data_arc_new.csv`**
- `key` — номер партии;
- `Начало нагрева дугой` — время начала нагрева;
- `Конец нагрева дугой` — время окончания нагрева;
- `Активная мощность` — значение активной мощности;
- `Реактивная мощность` — значение реактивной мощности.

**Файл `data_bulk_new.csv`**
- `key` — номер партии;
- `Bulk 1` … `Bulk 15` — объём подаваемого материала.

**Файл `data_bulk_time_new.csv`**
- `key` — номер партии;
- `Bulk 1` … `Bulk 15` — время подачи материала.

**Файл `data_gas_new.csv`**
- `key` — номер партии;
- `Газ 1` — объём подаваемого газа.

**Файл `data_temp_new.csv`**
- `key` — номер партии;
- `Время замера` — время замера;
- `Температура` — значение температуры.

**Файл `data_wire_new.csv`**
- `key` — номер партии;
- `Wire 1` … `Wire 15` — объём подаваемых проволочных материалов.

**Файл `data_wire_time_new.csv`**
- `key` — номер партии;
- `Wire 1` … `Wire 15` — время подачи проволочных материалов.

Во всех файлах столбец `key` содержит номер партии. В файлах может быть несколько строк с одинаковым значением `key`: они соответствуют разным итерациям обработки.

```python
def view_on_data(data):
    '''
    Input:
    - data (DataFrame): исходный датафрейм с данными
    
    Output:
    - None
    
    Description:
    Эта функция помогает провести первичный осмотр данных с помощью вывода
    первых 5 строк датасета, общей информации о датасете методом info() и основных описательных статистик.
    '''
    print('Информация о датасете')
    data.info()
    print('-'*60)
    print('Первые 5 строк датасета')
    display(data.head())
    print('-'*60)
    print('Описательные статистики показателей датасета')
    display(data.describe())
    print('-'*60)
    print('Кол-во строк-дубликатов в данных:', data.duplicated().sum())
```

```python
def num_col_analysis(data, column, label, xlim=None, bins=25, boxplot=True, stats=True):
    '''
    Input:
    - data (DataFrame): исходный датафрейм с данными;
    - column (String): название столбца для анализа;
    - label (String): подпись к гистограмме и диаграмме размаха, описывающая показатель;
    - lim (Tuple): ограничение по оси x. None по умолчанию;
    - bins (Integer): число корзин для построения гистограммы. По умолчанию bins=25;
    - boxplot(Boolean): Строить ли диаграмму размаха. True по умолчанию;
    - stats (Boolean): Выводить ли описательные статистики. True по умолчанию.
    
    Output:
    - None
    
    Description:
    Функция строит гистограмму распределения и диаграмму разброса выбранного столбца в датасете
    c одновременным выводом описательных статистик по показателю.
    '''

    # Задаем график и сетку
    fig = plt.figure(figsize=(12,9))
    fig.suptitle(column)
    gs = GridSpec(2, 1, height_ratios=[3.5, 1])
    
    # Добавляем гистограмму
    ax1 = fig.add_subplot(gs[0])
    ax1.hist(data[column], bins=bins)
    ax1.set_title('Гистограмма распределения показателя',y=1)
    #ax1.set_xlabel(label)
    ax1.set_ylabel('Кол-во наблюдений')
    ax1.set_xlim(xlim)
    
    # Добавляем диаграмму размаха
    if boxplot==True:
        ax2 = fig.add_subplot(gs[1])
        ax2.boxplot(data[column],vert=False)
        ax2.set_title('Диаграмма размаха', y=1)
        ax2.set_xlabel(label)
        ax1.set_xlim(xlim)
    else:
        pass

    plt.show();
    
    # Вывод описательной статистики о показателе
    if stats==True:
        temp = data.describe().T
        print('Описательная статистика показателя')
        return temp.loc[temp.index==column]
    else: 
        pass
```

```python
def histograms(data):
    '''
    Input:
    - data (DataFrame): исходный датафрейм с данными;
    
    Output:
    - None
    
    Description:
    Функция строит гистограммы распределения сразу для всех столбцов таблицы, кроме ключа партии.
    '''
    
    for column in data.drop(columns=['key']).columns:
        sns.set(rc={"figure.figsize": (20, 5)})
        sns.set_style('whitegrid')
        hist = sns.histplot(data=data, x=column, kde=True,  color='blue', alpha = 0.6, discrete=True)
        hist.set_ylabel('Кол-во наблюдений')
        hist.set_title('Гистограмма распределения показателя ' + column)
        plt.show()
```

## Объединение данных

### Подготовка таблиц и добавление новых признаков

Среди имеющихся датасетов есть несколько таблиц, которые необходимо агрегировать перед тем, как объединять все данные в единую таблицу. Это данные о мощности электродов `data_arc`, а также данные о замерах температуры `data_temp`. 

В этих таблицах также содержатся столбцы с временем в формате datetime, который вероятно будет не очень полезным при моделировании: в процессе плавки стали не важно, в какой день ее плавили, скорее важно время нагрева стали. Поэтому в таблицы с данными мы можем добавить столбцы с продолжительностью времени нагрева, а также временем, прошедшим между первым замером температуры и последним.

**Данные о мощности электродов (data_arc)**

В данных о мощности электродов нам нужно получить агрегированные значения мощности по каждой партии. До агрегирования найдем время нагрева стали в каждой итерации как разнице между временем конца нагрева и временем начала нагрева.

```python
# Вычислим продолжительность нагрева стали в каждой итерации
data_arc['heating_time'] = (data_arc['arc_heating_end']-data_arc['arc_heating_start']).dt.seconds

# Выведем первые 5-строк для проверки
data_arc.head()
```

Проведем агрегирование активной, реактивной мощностей и времени нагрева. При агрегировании не будем пока избавляться от времени начала и окончания нагрева, а найдем первое и последнее время нагрева. Одновременно вычислим количество итераций нагрева каждой из партий стали.

```python
# Создадим агрегированную таблицу с данными о нагреве каждой партии стали
data_arc_agg = data_arc.groupby('key').agg({'arc_heating_start':'min',
                                            'arc_heating_end':'max',
                                            'active_power':'sum',
                                            'reactive_power':'sum',
                                            'heating_time':['sum','count']
                                           })
# Зададим названия для колонок агрегиованной таблицы
data_arc_agg.columns = ['arc_heating_start', 'arc_heating_end', 'total_active_power', 'total_reactive_power', 'total_heating_time', 'n_heatings']

# Проверка
print(data_arc_agg.shape)
data_arc_agg.head()
```

Добавим в таблицу с данными о мощности электродов столбец с полной мощностью, которая вычисляется по формуле:\
*Полная мощность = √ (Активная мощность^2 + Реактивная мощность^2)*

```python
data_arc_agg['apparent_power'] = (data_arc_agg['total_active_power']**2 + data_arc_agg['total_reactive_power']**2)**0.5
data_arc_agg.head()
```

**Данные о температуре (data_temp)**

В данных о замерах температуры нам необходимо выделить первый и последний замер температуры и избавиться от всех промежуточных замеров (их нельзя использовать в качестве признаков из-за утечки данных).

```python
# Отсортируем таблицу с температурой по возрастанию даты
data_temp = data_temp.sort_values(by=['key','measurement_time'])
data_temp.head(15)
```

```python
# Найдем первое измерение температуры по каждой партии с помощью pivot_table
data_temp_first = data_temp.pivot_table(index='key', 
                                        values=['measurement_time', 'temperature'],
                                       aggfunc={'measurement_time':'first', 'temperature':'first'})

# Поменяем название колонок
data_temp_first.columns = ['measurement_time_first', 'temperature_first']

print(data_temp_first.shape)
data_temp_first.head()
```

```python
# Найдем последнее измерение температуры по каждой партии с помощью pivot_table
data_temp_last = data_temp.pivot_table(index='key', 
                                        values=['measurement_time', 'temperature'],
                                       aggfunc={'measurement_time':'last', 'temperature':'last'})

# Поменяем название колонок
data_temp_last.columns = ['measurement_time_last', 'temperature_last']

print(data_temp_last.shape)
data_temp_last.head()
```

```python
# Объединим данные по первому и последнему замеру в одну таблицу
data_temp_agg = data_temp_first.join(data_temp_last, how='inner')

# Проверка
print(data_temp_agg.shape)
data_temp_agg.head()
```

Теперь мы можем вычислить время между последним и первым замером температуры сплава. Время будет указано в секундах.

```python
# Вычислим время между первым и последним замером температуры
data_temp_agg['alloy_preparation_time'] = (data_temp_agg['measurement_time_last']-data_temp_agg['measurement_time_first']).dt.seconds

# Выведем первые 5-строк для проверки
print(data_temp_agg.shape)
data_temp_agg.head()
```

### Объединение датасетов

Теперь можем перейти непосредственно к объединению данных. Присоединим к датасету с целевой переменной `data_temp_agg` датасеты: 
- `data_arc_agg` - содержащий агрегированные  данные о нагреве сплава и мощности электродов;
- `data_bulk` - содержащий данные о подаче сыпучих материалов;
- `data_wire` - содержащий данные о подаче проволоки;
- `data_gas` - содержащий данные о продувке газом.

В датасетах, в которых мы производили агрерировние данных, номер партии `key` выступает индексом. Установим ключ key в качестве индекса в остальных таблицах.

```python
# Установим key в качестве индекса
data_bulk = data_bulk.set_index('key')
data_wire = data_wire.set_index('key')
data_gas = data_gas.set_index('key')

# Проверка
display(data_bulk.head())
display(data_wire.head())
display(data_gas.head())
```

```python
# Объединяем датасеты по ключу партии с использоваением типа слияния inner
data_all = data_temp_agg.join([data_arc_agg, data_bulk, data_wire, data_gas], how='inner')

# Выведем размерность объединенной таблицы
print('Размерность объединенной таблицы:', data_all.shape)

# Выведем последние строки полученной таблицы
display(data_all.tail(10))

# Выведем информацию о полученной таблице
data_all.info()
```

Это значительное количество признаков, многие из которых, как мы отмечали выше ввиду малого количества данных врядли будут вносить значительный вклад в предсказание температуры сплава. Поэтому следующим этапом произведем удаление признаков, которые определенно являются лишними.

```python
num_cols = X_train.columns

# Создаем паплайн подготовки признаков
data_preprocessor = ColumnTransformer([
                                        ('num', StandardScaler(), num_cols),
                                       ],
                                     remainder = 'passthrough')
```

## Обучение моделей машинного обучения

Как мы отметили ранее, линейных зависимостей между таргетом и признаками не обнаружено, поэтому мы будем использовать дерево решений в качестве простой модели. В качестве более сложных моделей обучим модели градиентного бустинга Light GBM и CatBoost.

Находить наилучшие параметры для каждой из моделей будем внутри пайплайна и с помощью GridSearhCV. Чтобы не повторять код запуска пайплайнов с различными моделями зададим соответствующую функцию.

```python
# Зададим функцию запуска пайплайна
def run_pipe(pipe_name, param_grid, X_train, y_train):
    
    # Ищем модель с наилучшими параметрами
    grid = GridSearchCV(pipe_name,
                       param_grid,
                       cv=5,
                       n_jobs=-1,
                       scoring='neg_mean_absolute_error')
    
    # Обучаем модель
    grid.fit(X_train, y_train)
    
    # Сохраняем в переменные параметры лучшей модели
    model = grid.best_estimator_
    params = grid.best_params_
    metric = - grid.best_score_
    
    return model, params, metric
```

```python
# Создаем пайплайн для дерева решений
pipe_decision_tree = Pipeline([
                       ('preprocessor', data_preprocessor),
                       ('models', DecisionTreeRegressor(random_state=RANDOM_STATE))
                     ])

# Создаем словарь для подбора гиперпараметров

param_grid = [

    # Словарь для модели дерева решений
    {
        'models':[DecisionTreeRegressor(random_state=RANDOM_STATE)],
        'models__max_depth':[2,5,7,9],
        'models__min_samples_split':range(2,6),
        'preprocessor__num':[StandardScaler(), MinMaxScaler(), 'passthrough']
    }
]

model_tree, params_tree, metric_tree = \
run_pipe(pipe_decision_tree, param_grid, X_train, y_train)

print('Параметры лучшей модели:', params_tree)
print('MAE по результатам кросс-валидации:', metric_tree)
```

```python
# Создаем пайплайн для lgbm
pipe_lgbm = Pipeline([
                       ('preprocessor', data_preprocessor),
                       ('models', LGBMRegressor(random_state=RANDOM_STATE, learning_rate=0.1))
                     ])

# Создаем словарь для подбора гиперпараметров

param_grid = [

    # Словарь для модели дерева решений
    {
        'models':[LGBMRegressor(random_state=RANDOM_STATE,learning_rate=0.1)],
        #'models__learning_rate': [0.01,0.1,0.3],
        'models__n_estimators':[30,50,100],
        'models__max_depth':[-1, 5, 8],
        'preprocessor__num':[StandardScaler(), MinMaxScaler(), 'passthrough']
    }
]

model_lgbm, params_lgbm, metric_lgbm = \
run_pipe(pipe_lgbm, param_grid, X_train, y_train)

print('Параметры лучшей модели:', params_lgbm)
print('MAE по результатам кросс-валидации:', metric_lgbm)
```

```python
# Создаем пайплайн для catboost
pipe_catboost = Pipeline([
                       ('preprocessor', data_preprocessor),
                       ('models', CatBoostRegressor(random_state=RANDOM_STATE, verbose=0, learning_rate=0.1))
                     ])

# Создаем словарь для подбора гиперпараметров

param_grid = [

    # Словарь для модели дерева решений
    {
        'models':[CatBoostRegressor(random_state=RANDOM_STATE, verbose=0, learning_rate=0.1)],
        'models__depth': [2, 4, 6],
        'models__n_estimators':[30, 50, 100],
        #'models__learning_rate': [0.01, 0.1, 0.5],
        'preprocessor__num':[StandardScaler(), MinMaxScaler(), 'passthrough']
    }
]

model_cat, params_cat, metric_cat = \
run_pipe(pipe_catboost, param_grid, X_train, y_train)

print('Параметры лучшей модели:', params_cat)
print('MAE по результатам кросс-валидации:', metric_cat)
```

Так, например, мы распролагали данными об объеме подаваемых материалов, но данных об объеме и составе изначального сплава у нас не было. Каким был объем сплава? Каково содержание железа? ### Проверка адекватности выбранной модели

Обучим модель DummyRegressor для проверки адекватности выбранной нами модели. Обучение модели осуществим сразу для двух стратегий: `mean` - использование средней в качестве прогноза, `median` - использование медианы в качестве прогноза.

```python
for strategy in ['median','mean']:
    dummy_model = DummyRegressor(strategy)
    dummy_model.fit(X_train, y_train)
    predictions_dummy = dummy_model.predict(X_test)
    MAE_dummy = mean_absolute_error(y_test, predictions_dummy)
    print('MAE для стратегии '+strategy, MAE_dummy)
```

Мы видим, что значение целевой метрики для обеих стратегий гороздо выше метрики не только выбранной модели, но и метрик всех обученных нами моделей. Это значит, что модель Dummy Regressor обладает хуже остальных предсказывает целевой признак. Мы можем считать выбранную модель адекватной.
